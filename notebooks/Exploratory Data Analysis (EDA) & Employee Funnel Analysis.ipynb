## Section 8 — Employee Lifecycle Funnel  
We map employees through stages: Hire → Onboard → Engage → Perform → Retain → Attrition.
This allows us to understand where major employee drop-offs occur.


```python
stage1 = len(df)
```

```python
stage2 = df[df['YearsAtCompany'] <= 1].shape[0]
```

```python
# Engaged Employees
stage3 = df[df['EngagementScore'] >= 3].shape[0]
```

```python
# Performing Employees
stage4 = df[df['PerformanceRating'] >= 3].shape[0]
```

```python
# Retained Employees
stage5 = df[(df['ManagerRating'] >= 3) & (df['OverTimeHours'] <= df['OverTimeHours'].median())].shape[0]
```

```python
#Attrition
stage6 = df[df['AttritionFlag'] == 1].shape[0]
```

```python
funnel = pd.DataFrame({
    'Stage': ['Hire','Onboard','Engage','Perform','Retain','Attrition'],
    'Count': [stage1,stage2,stage3,stage4,stage5,stage6]
})

funnel['ConversionRate'] = funnel['Count'] / stage1
funnel['DropOffRate'] = 1 - funnel['ConversionRate']
funnel
```

## Section 9 — Funnel Visualization  
This funnel visually highlights bottlenecks such as engagement and retention.


```python
plt.figure(figsize=(12,5))
sns.barplot(x='Stage', y='Count', data=funnel, palette="Blues_d")
plt.title("Employee Lifecycle Funnel")
plt.show()
```
