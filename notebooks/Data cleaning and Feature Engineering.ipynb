```python
df['Department'] = df['Department'].replace({'Rnd': 'R&D'})
```

```python
df['Gender'] = df['Gender'].str.strip().str.title()
df['Education'] = df['Education'].str.strip().str.title()
```

```python
cols_to_check = [
    'MonthlyIncome',
    'DistanceFromHome',
    'OverTimeHours',
    'TrainingHoursLastYear',
    'Age',
    'TotalWorkingYears',
    'YearsAtCompany',
    'YearsInCurrentRole'
]
```

```python
for col in cols_to_check:
    
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    
    print(f"\nColumn: {col}")
    print(f"Lower Bound: {lower_bound}")
    print(f"Upper Bound: {upper_bound}")
    print(f"Outliers Found: {len(outliers)}")
```

```python
df['AttritionFlag'] = df['Attrition'].map({'Yes': 1, 'No': 0})
```

# Feature Engineering

```python
df['EngagementScore'] = df[['JobSatisfaction','WorkLifeBalance','ManagerRating']].mean(axis=1)
```

```python
#Used to identify burnout
df['OvertimeCategory'] = pd.qcut(

    df['OverTimeHours'],
    q=3,
    labels=['Low','Medium','High']
)
```

```python
#Early leavers (0-1) usually have highest attrition
df['TenureBucket'] = pd.cut(
    df['YearsAtCompany'],
    bins=[0,1,4,9,50],
    labels=['0-1','2-4','5-9','10+']
)
```

```python
df['IsEarlyJoiner'] = (df['YearsAtCompany'] <= 2).astype(int)
```

```python
#Recognize managers with higher churn
df['ManagerLow'] = (df['ManagerRating'] <= 3).astype(int)
```

```python
#Detect underpaid employees
df['CompensationRatio'] = df['MonthlyIncome'] / df.groupby('JobRole')['MonthlyIncome'].transform('median')
```

```python
#Identifies overworked employees
df['WorkStrainScore'] = (
    (df['OverTimeHours'] / df['OverTimeHours'].max()) * 0.5 +
    (df['DistanceFromHome'] / df['DistanceFromHome'].max()) * 0.3 +
    (1 - df['WorkLifeBalance'] / 5) * 0.2
)
```

```python
df.to_csv("clean_hr_data.csv", index=False)
```
